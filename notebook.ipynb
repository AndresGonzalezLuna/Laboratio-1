### Introducción

En las finanzas siempre ha existido un debate acerca de las estrategias de inversión. Hay gente que opina que es mejor tener un enfoque activo, en el que constantemente estás vendiendo y comprando acciones dependiendo de como se mueva el mercado. Hay gente que opina que es mejor tener una estrategia pasiva, lo cuál significa comprar unas acciones y no alterar la posición hasta que se vende. Muchos financieros son portfolio managers, los cuales su trabajo es manejar el dinero de otras personas. Hay mucho dinero en las comisiones que se cobran por manejar este dinero y por las comisiones en las diferentes transacciones. Si se tuviera una postura pasiva para la inversión, no sería necesario andar pagando constantemente estas comisiones. Lo importante es usar la estrategia que maximice tu dinero y es lo que vamos a tratar de ver aqui.

### Inversión pasiva

El objetivo era obtener un DataFrame que mostrara el performance de nuestro portafolio de $1m con una estrategia de inversión pasiva. Utilizamos el archivo más antiguo ("NAFTRAC_20210129") como nuestra base. Sacamos el número de acciones que teníamos por cada Ticker:    


$$(Inversion Inicial \times Peso (%))\divPrecio$$ 
  
  En base en este número de acciones por Ticker que compramos en este inicio, pudimos utilizar el precio de cierre de mes para ver de que tamaño era nuestra posición en ese momento. Si sumamos todas nuestras posiciones para ese día, nos da el valor de nuestro portafolio. 

```python
# Importamos data del archivo inicial con las ponderaciones
from data import data
```

```python
#Transformamos
df = functions.transform_data(data = data, monto_inv_inicial = 1000000, comision=0.00125)
```

Utilizamos las fechas de cierre para solo agregar estas fechas a un diccionario, se podría decir que es donde filtramos.

```python
#%% Sacamos las fechas de cierre de mes de los archivos

fechas_cierre = functions.date_checker('files/')[0]
fecha_inicial = functions.date_checker('files/')[1]
```

Había algunos tickers con diferencias a los que salían a Yahoo, principalmente contenían un asterisco (*) que quitamos. 

```python
#%% Limpiamos los tickers

tickers = functions.ticker_cleaner(df)
```

Creamos un diccionario para cada Ticker que contenía como keys las fechas de los días de fin de mes, y los valores de la posición. La comision solo se cobra en la fecha incial. Cada uno de estos diccionarios estaba dentro de un diccionario mayor que acabaría conteniendo todos los valores de nuestras posiciones. Creamos un DataFrame con este diccionario y sacamos el rendimiento de nuestro portafolio y el rendimiento acumulado del mismo

```python
#%% Descargamos los datos para cada activo, y agregamos los valores a la fecha de cierre a un diccionario de diccionarios

diccionario_tickers = functions.dictionary_maker(tickers=tickers, fecha_inicial=fecha_inicial, df=df, fechas_cierre=fechas_cierre, comision=0.00125)

#%% Convertimos el diccionairo a un df para hacerlo como se pide en el proyecto

df_pasiva = functions.final_df_maker(diccionario_tickers, fecha_inicial)
```
